```python
from seaborn.matrix import clustermap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.lmplot('age','income',data=df,fit_reg=False,size=5)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaled_df=scaler.fit_transform(df[['age','income']])

scaled_df[:5]
```

```python
from sklearn.cluster import KMeans
clusters=KMeans(3)

clusters.fit(scaled_df)

df['Cluster']=clusters.labels_
```

```python
markers=['*','^','+']
sns.lmplot('age','income',data=df,hue='Cluster',fit_reg=False,markers=markers,size=5)
```
